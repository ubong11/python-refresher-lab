```python
# =========================
# FUNDAMENTALS OF PYTHON
# =========================

# Data Types
age = 25
height = 5.9
name = "Henry Falaiye"
is_student = True

print(age, height, name, is_student)

# Conditional Statements
score = 75

if score >= 70:
    print("Grade: A")
elif score >= 50:
    print("Grade: B")
else:
    print("Grade: C")

# Function and indentation
def check_even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check_even_or_odd(10))
```

```python
# =========================
# WORKING WITH DATA
# =========================

# Lists
numbers = [1, 2, 3, 4, 5]
numbers.append(6)
numbers.remove(2)

print("Updated list:", numbers)

# Loops
for num in numbers:
    print("Number:", num)

# range()
for i in range(1, 11):
    print("Range value:", i)

# File I/O
with open("sample.txt", "w") as file:
    file.write("This is a sample file for Henry Falaiye.")

with open("sample.txt", "r") as file:
    content = file.read()

print("File content:", content)
```

```python
# =========================
# ADVANCED TOPICS
# =========================

# Pickle
import pickle

data = {"name": "Henry", "age": 25}

with open("data.pkl", "wb") as file:
    pickle.dump(data, file)

with open("data.pkl", "rb") as file:
    loaded_data = pickle.load(file)

print("Pickle data:", loaded_data)
```

```python
import h5py
import numpy as np

data = np.arange(10)

with h5py.File("data.h5", "w") as file:
    file.create_dataset("dataset_1", data=data)

with h5py.File("data.h5", "r") as file:
    dataset = file["dataset_1"][:]

print("HDF5 data:", dataset)
```

```python
import sqlite3

conn = sqlite3.connect("students.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER
)
""")

cursor.execute("INSERT INTO students (name, age) VALUES (?, ?)", ("Henry", 25))
conn.commit()

cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

for row in rows:
    print("DB Record:", row)

conn.close()
```
